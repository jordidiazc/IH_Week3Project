```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

# Crear un DataFrame de ejemplo con datos de dispersión
data = {
    'Temperature': ["a", "a", "b", "b", "b", "b", "c"],
    'Ice_Cream_Sales': [100, 120, 150, 180, 220, 260, 300]
}
df = pd.DataFrame(data)

# Crear un gráfico de dispersión
fig, ax = plt.subplots()
ax.barh(df['Temperature'], ax=ax, color='red')

# Añadir títulos y etiquetas
ax.set_title('Ice Cream Sales vs Temperature')
ax.set_xlabel('Temperature (°C)')
ax.set_ylabel('Ice Cream Sales')

# Mostrar el gráfico
plt.show()
```

```python
lista1 = ['apple', 'banana', 'apple', 'orange', 'banana', 'apple', 'kiwi']

# Contar las ocurrencias de cada elemento en la lista
conteo = Counter(lista1)

# Obtener las listas para los ejes X e Y
elementos = list(conteo.keys())
frecuencias = list(conteo.values())

# Crear el gráfico de barras
fig, ax = plt.subplots()
ax.barh(elementos, frecuencias, color='skyblue')
```
